```python
#15. Elabore um algoritmo que solicite N números reais e quando o usuário informar o valor nulo 0 (zero) o programa ordene e mostre todos os números informados de forma crescente.

numeros = []

print("Digite quantos números reais desejar.")
print("Para finalizar e ordenar, digite o número 0 (zero).")

while True:
    numero_lido = float(input("Digite um número: "))

    if numero_lido == 0:
        break

    numeros.append(numero_lido)

if numeros:
    numeros.sort()

    print(*numeros, sep='\n')
else:
    print("\nNenhum número foi informado para ordenação.")
```

```python
idades = []

print("--- Cadastro de Idades dos Alunos ---")

while True:
    idade_str = input("\nDigite a idade do aluno: ")

    if idade_str.isdigit() and int(idade_str) > 0:
        idade_atual = int(idade_str)
        idades.append(idade_atual)
    else:
        print("Entrada inválida. Por favor, digite um número inteiro positivo.")
        continue # Volta para o início do loop

    # Loop para perguntar se deseja continuar
    while True:
        continuar = input("Deseja informar a idade do próximo aluno? (S/N): ").upper()
        if continuar == 'S' or continuar == 'N':
            break # Sai do loop S/N
        else:
            print("Opção inválida. Por favor, digite S para Sim ou N para Não.")

    # Se a resposta for 'N', sai do loop principal
    if continuar == 'N':
        break

# --- Se a lista não estiver vazia, calcula as estatísticas ---
if idades:
    # Ordena a lista para calcular min/max e mediana
    idades.sort()

    # 1. Mais novo e Mais velho
    mais_novo = idades[0]
    mais_velho = idades[-1] # -1 pega o último item da lista

    # 2. Contagem de maiores e menores de 18
    maiores_de_18 = 0
    ate_18 = 0
    indice = 0
    total_alunos = len(idades)

    while indice < total_alunos:
        idade = idades[indice]
        if idade > 18:
            maiores_de_18 += 1
        else:
            ate_18 += 1
        indice += 1

    # 3. Média Aritmética
    soma_idades = sum(idades)
    media = soma_idades / total_alunos

    # 4. Cálculo da Mediana (O BLOCO QUE FALTAVA)
    mediana = 0
    meio = total_alunos // 2 # Divisão inteira para achar o índice do meio

    if total_alunos % 2 == 1:
        # Se o total for ÍMPAR, a mediana é o valor do meio
        mediana = idades[meio]
    else:
        # Se o total for PAR, a mediana é a média dos dois valores centrais
        mediana = (idades[meio - 1] + idades[meio]) / 2

    # --- Saída dos Resultados ---
    print("\n--- Estatísticas da Turma ---")
    print(f"Idade do aluno mais novo: {mais_novo} anos")
    print(f"Idade do aluno mais velho: {mais_velho} anos")
    print(f"Alunos com mais de 18 anos: {maiores_de_18}")
    print(f"Alunos com até 18 anos (inclusive): {ate_18}")
    print(f"Média de idade da turma: {media:.2f} anos")
    print(f"Mediana das idades: {mediana} anos")

else:
    print("\nNenhuma idade foi informada para gerar estatísticas.")
```
