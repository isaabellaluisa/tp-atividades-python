# **Sistema Simplicado de Cadastro de Aluno**

**Lista de Dicion√°rios para simular um banco de dados de alunos.**

```python
alunos_db = [
  {"matricula": "2025001", "nome": "Isabella Lu√≠sa", "curso": "Analise de Desenvolvimento de Sistemas"},
  {"matricula": "2025002", "nome": "Larissa", "curso": "Enfermagem"},
  {"matricula": "2025003", "nome": "Pablo", "curso": "Analise de Desenvolvimento de Sistemas"},
  {"matricula": "2025004", "nome": "Amon", "curso": "Analise de Desenvolvimento de Sistemas"},
]
print("1. Sistema simplificado de cadastro de alunos")
print("-" * 40)

print("-> Adicionando novo aluno...")
novo_aluno = {"matricula": "2025001", "nome": "Isabella Lu√≠sa", "curso": "Analise de Desenvolvimento de Sistemas"}
alunos_db.append(novo_aluno)
print(f"Aluno {novo_aluno['nome']} adicionado com sucesso.")

matricula_busca = "2025001"

aluno_encontrado = next((aluno for aluno in alunos_db if aluno["matricula"] == matricula_busca), None)

if aluno_encontrado:
    print(f"\nAluno encontrado: (matricula: {matricula_busca}):")
    print(f"Nome: {aluno_encontrado['nome']}, curso: {aluno_encontrado['curso']}")
else:
    print(f"\nAluno com matr√≠cula {matricula_busca} n√£o encontrado.")


matricula_update = "2025001"
for aluno in alunos_db:
  if aluno["matricula"] == matricula_update:
    aluno["curso"] = "Engenharia de Software"
    print(f"\nCurso do aluno {aluno['nome']} atualizado para: {aluno['curso']}")
    break
print("-" * 40)

print(alunos_db)
```

# **Gerenciamento de Coordenadas Geogr√°ficas**

**Lista de Tuplas para armazenar coordenadas, garantindo que a latitude e longitude de um local sejam imut√°veis ap√≥s o registro.**

```python
locais = [
    (-41.2865, 174.7762, "Nova Zel√¢ndia"),   # Wellington
    (-6.2088, 106.8456, "Indon√©sia"),        # Bali
    (-15.7975, -47.8919, "Bras√≠lia")         # Bras√≠lia
]

print("\nüìç Sistema de Coordenadas Personalizado")
print("=" * 50)

# Inserindo um novo ponto
novo_local = (-36.8485, 174.7633, "Auckland")  # Outro ponto da Nova Zel√¢ndia
locais.append(novo_local)
print(f" Novo local inclu√≠do: {novo_local[2]}")

# Exibindo a √∫ltima coordenada adicionada
print("\n √öltimo ponto cadastrado:")
lat, lon, nome = locais[-1]
print(f"Local: {nome} | Lat: {lat} | Lon: {lon}")

# Listando todas as coordenadas registradas
print("\n Lista de locais armazenados:")
for coordenada in locais:
    print(f"- {coordenada[2]} ‚Üí Latitude: {coordenada[0]}, Longitude: {coordenada[1]}")
```

# **Cat√°logo de Produtos com Busca R√°pida**

**Dicion√°rio onde a chave √© o ID do produto, permitindo uma busca extremamente r√°pida do objeto do produto.**

```python
catalogo = {
    "P001": {"nome": "XJ6", "estoque": 30, "preco": 46000.00},
    "P002": {"nome": "XT660", "estoque": 40, "preco": 53000.00},
    "P003": {"nome": "S1000", "estoque": 10, "preco": 96000.00}
}

print("\n Cat√°logo de Motos com Busca r√°pida")
print("-" * 40)

novo_moto_id = "P004"
catalogo[novo_moto_id] = {"nome": "StreetFigther", "estoque": 70, "preco": 90000.00}
print(f"Moto '{catalogo[novo_moto_id]['nome']}' adicionada ao cat√°logo.")

id_busca = "P002"
if id_busca in catalogo:
    moto = catalogo[id_busca]
    print(f"\nDetalhes da moto {id_busca}:")
    print(f"Nome: {moto['nome']}, Pre√ßo: R${moto['preco']:.2f}")

else:
  print(f"\nMoto com ID {id_busca} n√£o encontrada.")

id_update = "P001"
catalogo[id_update]['estoque'] -= 5
print(f"\nEstoque da {catalogo[id_update]['nome']} atualizado. Novo Estoque: {catalogo[id_update]['estoque']}")
print("-" * 40)
```

# **Agrupamento de Vendas por Regi√£o**

**Dicion√°rios Aninhados e Tuplas para estruturar dados mais complexos.**

```python
dados_vendas = {
    "SUDESTE": {
        "total_anual": 150000.00,
        "vendas_detalhes": [
            ("V001", 5000.00),
            ("V002", 12000.00),
            ("V003", 8000.00)
        ]
    },
    "SUL": {
        "total_anual": 95000.00,
        "vendas_detalhes": [
            ("V004", 6500.00),
            ("V005", 10000.00)
        ]
    }
}

print("\nAgrupamento de Vendas por Regi√£o")
print("-" * 40)

regiao = "SUDESTE"
nova_venda = ("V006", 9500.00)

dados_vendas[regiao]["vendas_detalhes"].append(nova_venda)
print(f"Nova venda {nova_venda[0]} adicionada na regi√£o {regiao}.")

vendas_sudeste = dados_vendas["SUDESTE"]["vendas_detalhes"]
soma_detalhes = sum(venda[1] for venda in vendas_sudeste)

print(f"\nSoma dos detalhes de vendas na regi√£o {regiao}: R${soma_detalhes:.2f}")
print(f"Total anual registrado: R${dados_vendas[regiao]['total_anual']:.2f}")

print("\nRegi√µes em an√°lise:")
for reg in dados_vendas.keys():
  print(f"- {reg}")
print("-" * 40)
```

# **Mapeamento de N√≠veis de Acesso e Permiss√µes**

**Dicion√°rio para mapear o n√≠vel de acesso (chave) para uma Tupla de permiss√µes (valor), garantindo que o conjunto de permiss√µes para cada n√≠vel seja fixo e imut√°vel.**

```python
permissoes = {
    "ADMIN": ("CREATE_USER", "READ_ALL", "UPDATE_ANY", "DELETE_ALL"),
    "MANAGER": ("READ_ALL", "UPDATE_OWN"),
    "VIEWER": ("READ_ALL",)
}

print("\n Mapeamento de N√≠veis de Acesso e Permiss√µes")
print("-" * 40)

nivel_acesso = "ADMIN"
perms_admin = permissoes.get(nivel_acesso)

if perms_admin:
  print(f"Permiss√µes do n√≠vel '{nivel_acesso}':")
  for perm in perms_admin:
    print(f"- {perm}")

else:
  print(f"N√≠vel de acesso '{nivel_acesso}' n√£o definido.")

perm_verificar = "DELETE_ALL"
if perm_verificar in permissoes["ADMIN"]:
  print(f"\nADMIN possui a permiss√£o: {perm_verificar}")
else:
  print(f"\nADMIN n√£o possui a permiss√£o: {perm_verificar}")

```
